```python
# saving dictionary of dataframes

def saver(dictex, label):
    for key, val in dictex.items():
        val.to_csv(label + "_data_{}.csv".format(str(key)))

    with open(label + "_keys.txt", "w") as f: #saving keys to file
        f.write(str(list(dictex.keys())))

def loader(keys, label):
    """Reading data from keys"""
    with open(keys, "r") as f:
        keys = eval(f.read())

    dictex = {}    
    for key in keys:
        dictex[key] = pd.read_csv(label + "_data_{}.csv".format(str(key)))

    return dictex
```

# RRBS simulated dilutions

```python
densityTablePath = "READtoMD.DT.RRBS.csv"
df = pd.read_csv(densityTablePath)

controls = [i for i in df.columns if "BuffyCoat" in i]
controlsDF = df[['numM', 'numU', 'MD'] + controls]
cases = [i for i in df.columns if "Ovarian_Cancer" in i]
casesDF = df[['numM', 'numU', 'MD'] + cases]
```



### average methylation values for "background" (controls)

```python
# meC / total C for all reads in a sample (so like EF but all reads)

BkGAveMeth = []
for control in controls:
    counts = controlsDF[control]  # read counts for each MD of the given sample
    total_c_per_row = (controlsDF['numU'].values +
                       controlsDF['numM'].values) * counts.values
    total_C = np.nansum(total_c_per_row)
    # compute average methylation (use to sort samples in plot)
    # (total number of meC's covered by reads / total C's covered by reads)
    ave_meC = np.nansum(controlsDF['numM'].values *
                        counts.values) / float(total_C)
    BkGAveMeth.append(ave_meC)
```

```python
def stringsToTuples(x):
    my_tuple = literal_eval(x)
    return my_tuple

def dilution(tumorFraction, depth, num_spikeins, controlsDF, controls, casesDF, cases):
    '''
    tumorFraction = fraction of tumor reads
    depth = total reads in spikein
    num_spikeins = number of spikein samples to make
    controlsDF = READtoMD.DT. of original control samples
    controls = list of control sample IDs
    casesDF = READtoMD.DT. of original case samples
    cases = list of case sample IDs
    '''

    # generate spikeIn samples
    spikeIn_df = pd.DataFrame(index=np.arange(depth))
    for i in np.arange(num_spikeins):
        case_sample = cases[i]  # pair each ov sample randomly with a WBC
        case_counts = casesDF[case_sample].values
        # frequency of each numU,numM MD combo
        case_probs = case_counts / float(np.nansum(case_counts))
        case_patterns = list(zip(casesDF['numU'], casesDF['numM']))
        case_patterns = [str(i) for i in case_patterns]

        control_sample = np.random.choice(
            a=controls, replace=False)  # randomly choose a wbc sample
        control_counts = controlsDF[control_sample].values
        control_probs = control_counts / float(np.nansum(control_counts))
        control_patterns = list(zip(controlsDF['numU'], controlsDF['numM']))
        control_patterns = [str(i) for i in control_patterns]

        # High density sampling spike-in:
        controlReadDepth = depth - (depth * tumorFraction)
        backgroundReads = pd.DataFrame(np.random.choice(a=control_patterns, size=(
            int(controlReadDepth), 1), p=control_probs))  # "background reads"

        caseReadDepth = (depth * tumorFraction)
        spikeInReads = pd.DataFrame(np.random.choice(a=case_patterns, size=(
            int(caseReadDepth), 1), p=case_probs))  # spikein reads

        # combine sampled read "patterns" (numU, numM)
        spikeIn = pd.concat([backgroundReads, spikeInReads], ignore_index=True)
        spikeIn_df = spikeIn_df.merge(
            spikeIn, right_index=True, left_index=True)  # add spikein to dataframe

    return spikeIn_df.applymap(stringsToTuples)
```

```python
def spikeInAveMeth(spikeIn_df):

    spikein_ave_vals = []

    for i in np.arange(len(spikeIn_df.columns)):
        sample = spikeIn_df.iloc[:, i].to_frame()
        # for each tuple, or pattern, pull out numU and numM information:
        sample['numU'] = sample.apply(lambda x: [float(y[0]) for y in x])
        sample['numM'] = [float(i[1]) for i in sample.iloc[:, 0].values]
        tot_c = np.nansum(sample['numU'].values + sample['numM'].values)
        meC = np.nansum(sample['numM'].values)
        ave_meC = meC/tot_c
        spikein_ave_vals.append(ave_meC)

    return spikein_ave_vals
```

```python
def computeBkgControlMDreadCounts(controlsDF, controls, mdc):
    '''
    Read counts here are technically EFs b/c need to weight by number of C's covered.
    '''
    counts = []
    # EF for given mdc for each control sample:
    for control in controls:
        total_c_per_row = (controlsDF['numU'] + controlsDF['numM']) * controlsDF[control]
        total_c = np.nansum(total_c_per_row)
        if mdc == 0.0:
            md_rows = controlsDF[controlsDF['MD'] > float(mdc)]
        else:
            md_rows = controlsDF[controlsDF['MD'] >= float(mdc)]
        row_c = (md_rows['numU'] + md_rows['numM']) * md_rows[control]
        sample_frac_c = row_c.sum()
        weighted_ef = sample_frac_c / float(total_c)
        counts.append(weighted_ef)
        
#     if mdc == 0.0:
#         selection = controlsDF[controlsDF['MD'] > float(mdc)]
#     else:
#         selection = controlsDF[controlsDF['MD'] >= float(mdc)]
#     counts = selection[controls].sum(axis=0, skipna=True).values.tolist()

    return counts
```

```python
def computeSpikeInMDreadCounts(spikeIn_df, mdc):
    '''
    read counts here are technically EFs b/c need to weight by number of C's covered.
    '''

    spikein_readCounts = []
    for i in np.arange(len(spikeIn_df.columns)):

        sample = spikeIn_df.iloc[:, i].to_frame()
        sample['numU'] = sample.apply(lambda x: [float(y[0]) for y in x])
        sample['numM'] = [float(i[1]) for i in sample.iloc[:, 0].values]
        sample['MD'] = sample['numM'].values / \
            (sample['numU'].values + sample['numM'].values)
        if mdc == 0.0:
            md_rows = sample[sample['MD'] > float(mdc)]
        else:
            md_rows = sample[sample['MD'] >= float(mdc)]

        frac_c = md_rows['numU'].sum() + md_rows['numM'].sum()
        tot_c = sample['numU'].sum() + sample['numM'].sum()
        weighted_sf = frac_c/tot_c
        spikein_readCounts.append(weighted_sf)

    return spikein_readCounts
```

```python
def ROC_stats(caseVals, controlVals):

    case_labels = [1 for i in caseVals]
    control_labels = [0 for i in controlVals]
    values = caseVals + controlVals
    labels = case_labels + control_labels
    fpr, tpr, thresholds = roc_curve(labels, values)
    roc_auc = auc(fpr, tpr)
    optimal_idx = np.argmax(tpr - fpr)
    optimal_threshold = thresholds[optimal_idx]

    df_roc = pd.DataFrame({'label': labels, 'values': values})
    true_positives = len(df_roc[(df_roc['label'] == 1) & (
        df_roc['values'] > optimal_threshold)].index)
    false_negatives = len(df_roc[(df_roc['label'] == 1) & (
        df_roc['values'] <= optimal_threshold)].index)
    true_negatives = len(df_roc[(df_roc['label'] == 0) & (
        df_roc['values'] <= optimal_threshold)].index)
    false_positives = len(df_roc[(df_roc['label'] == 0) & (
        df_roc['values'] > optimal_threshold)].index)
    sensitivity = float(true_positives) / \
        (float(true_positives) + float(false_negatives))
    specificity = float(true_negatives) / \
        (float(true_negatives) + float(false_positives))

    return roc_auc, sensitivity, specificity
```

```python
def RRBS_sim(dilutions, depth, num_spikeins, controlsDF, controls, casesDF, cases, MDcutoffs):
    '''
    dilutions = list of dilutions to make (as % tumor DNA fraction)
    depth = read depth for each sample (total reads in a spikein)
    num_spikeins = # of case spikeins to make aka random pairings with controls, per eac dilution
    controlsDF = READtoMD.DT. of original control samples
    controls = list of control sample IDs
    casesDF = READtoMD.DT. of original case samples
    cases = list of case sample IDs
    MDcutoffs = list of MD cutoff values to use (0.0, 0.05, 0.10, ..., 1.0)
    '''
    print('beginning simulation...')

    bestMDAUCsEachDil = []
    aveMethAUCsEachDil = []

    each_dilution_MDaucs = []
    each_dilution_MDsenss = []
    each_dilution_MDspecs = []

    for tumorfrac in dilutions:

        spikeIn_df = dilution(tumorfrac, depth, num_spikeins,
                              controlsDF, controls, casesDF, cases)
        spikein_AveMeth = spikeInAveMeth(spikeIn_df)
        aveAUC, aveSens, aveSpec = ROC_stats(spikein_AveMeth, BkGAveMeth)
        aveMethAUCsEachDil.append(aveAUC)

        # values for each md, for a given dilution
        md_aucs = []
        md_senss = []
        md_specs = []

        for mdc in MDcutoffs:

            controlMDcounts = computeBkgControlMDreadCounts(
                controlsDF, controls, mdc)
            caseMDcounts = computeSpikeInMDreadCounts(spikeIn_df, mdc)
            mdAUC, mdSens, mdSpec = ROC_stats(caseMDcounts, controlMDcounts)
            md_aucs.append(mdAUC)
            md_senss.append(mdSens)
            md_specs.append(mdSpec)

        best_auc_idx = np.argmax(md_aucs)
        best_auc = md_aucs[best_auc_idx]
        best_MDC = MDcutoffs[best_auc_idx]

        bestMDAUCsEachDil.append(best_auc)
        each_dilution_MDaucs.append(md_aucs)
        each_dilution_MDsenss.append(md_senss)
        each_dilution_MDspecs.append(md_specs)
    
    indices = [str(i) for i in dilutions]
    dilutionsMD_aucDF = pd.DataFrame(
        each_dilution_MDaucs, columns=MDcutoffs, index=indices)
    dilutionsMD_sensDF = pd.DataFrame(
        each_dilution_MDsenss, columns=MDcutoffs, index=indices)
    dilutionsMD_specDF = pd.DataFrame(
        each_dilution_MDspecs, columns=MDcutoffs, index=indices)

    print('simulation done.')

    return bestMDAUCsEachDil, aveMethAUCsEachDil, dilutionsMD_aucDF, dilutionsMD_sensDF, dilutionsMD_specDF
```
